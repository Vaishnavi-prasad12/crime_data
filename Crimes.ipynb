```python
import pandas as pd

# Load dataset
df = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert 'Date Reported' to datetime format
df["Date Rptd"] = pd.to_datetime(df["Date Rptd"], errors="coerce")

# Sort by Date Reported (most recent first)
df_sorted = df.sort_values(by="Date Rptd", ascending=False)

# Keep only the last 20,000 rows
df_recent = df_sorted.head(20000)

# Save the reduced dataset
df_recent.to_csv("recent_crime_dataset1.csv", index=False)
print("Dataset reduced to recent 20,000 crimes and saved!")


```
